```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# Check the percentile distribution of train, test, and validation samples of the sensitivity truth

def check_quarter_percentile(filename):
    df1 = pd.read_csv(filename, delimiter="\t", header=None)[2]
    plt.figure()
    df1.hist()
    plt.title(filename)
    df1[df1>=0.75] = 1
    df1[df1<=0.25] = 0
    print df1.value_counts().iloc[[0,1]]
    print df1.value_counts().iloc[[0,1]]/df1.shape[0]

all_values = "../psl/data/first_model/sensitive_truth.txt"
train = "../psl/data/first_model/seed0/cross_val/sensitive_truth_train.txt"
test = "../psl/data/first_model/seed0/test/sensitive_truth_test.txt"


check_quarter_percentile(all_values)
check_quarter_percentile(train)
check_quarter_percentile(test)

for fold in range(1, 6):
    for kind in ["train", "val"]:
        filename = "../psl/data/first_model/seed0/cross_val/5fold/fold{0}_{1}.txt".format(fold, kind)
        check_quarter_percentile(filename)
```
